```python
import pandas as pd
```

```python
content = pd.read_csv('Content_cleaned.csv')    
reactions = pd.read_csv('Reactions_cleaned.csv')
reaction_types = pd.read_csv('ReactionTypes_cleaned.csv')
```

```python
print(content.shape)
content.head()
```

```python
print(reactions.shape)
reactions.head()
```

```python
print(reaction_types.shape)
reaction_types.head()
```

```python
# left outer join on 'reactions' with 'reaction_types' on column ['Reaction Type']
merged_results = pd.merge(reactions, reaction_types, on='Reaction Type', how='left')
merged_results=merged_results.drop(columns=["Unnamed: 0_y"])

print(merged_results.shape)
merged_results.head()
```

```python
# left outer join on 'merged_results' with 'content' on 'Content ID'

merged_results = pd.merge(merged_results, content, on='Content ID', how='left')
merged_results=merged_results.drop(columns=["Unnamed: 0"])
merged_results = merged_results.rename(columns={"Unnamed: 0_x": "Index"})

print(merged_results.shape)
merged_results.head()
```
