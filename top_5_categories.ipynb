```python
import pandas as pd
```

```python
reactions = pd.read_csv("merged_cleaned_datasets.csv")
reactions.head()
```

```python
gb_category = reactions.groupby('Category')
avg_category_scores = gb_category['Score'].sum().sort_values(ascending=False).reset_index()
avg_category_scores
```

```python
# select just the top 5
top_5_categories = avg_category_scores[0:5]

# enumerate the ranking
top_5_categories['Rank'] = range(1,6)

# move the rank to the leftmost side of the dataframe
top_5_categories = top_5_categories.reindex(columns=['Rank', 'Category', 'Score'])
```

# Result
